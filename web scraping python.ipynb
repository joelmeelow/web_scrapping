```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

  
#looping through the pages
for i in range(1, 3):   
    #making a get request to the url and storing the response in a varaible, content
    content = requests.get("https://medplusnig.com/11-multivitamin?page={}".format(i)).text
    #using beautiful soup and Html paser to extract the content
    
    soup = BeautifulSoup(content, 'html.parser')
    #using the find_all method to extract h4 tags with class card-title and storing it in a variable, jobs
    jobs = soup.find_all('h4', class_="card-title")
    #initializing the output variables to an empty list
    drugname = []
    drugprice = []
    
    #looping through all the child components of jobs extracting key information  and appending them to the appropriate empty list

    for job in jobs:
        try:

            drugname.append(job.text[10:57]).str.strip()
        except:
            drugname.append(np.nan)
        try:
            drugprice.append(job.span.text.replace(",", "")).str.strip()
        except:
            drugprice.append(np.nan)

        
#creating a dictionary to store the list
result = {"drug_name": drugname, "drug_price": drugprice}

#inserting result in a Dataframe       
df = pd.DataFrame(result)
#save the csv
df.to_csv('medpluslist4')



```
